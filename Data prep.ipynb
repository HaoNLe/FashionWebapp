```python
labeled_data = []
labels = []

for key, value in Item.items():
    catLists = value[b'categories']
    tempSet = set()
    for catList in catLists:
        for cat in catList:
            tempSet.add(cat)
    
    gender = 2
    piece = 3
    for cat in tempSet:
        if cat in MALE_CATS:
            gender = 0
        elif cat in FEMALE_CATS:
            gender = 1
        elif cat in TOP_CATS:
            piece = 0
        elif cat in BOTTOM_CATS:
            piece = 1
        elif cat in SHOE_CATS:
            piece = 2
    
    if gender == 0 and piece == 0:
        labels.append([1, 0, 0, 0, 0, 0])
    elif gender == 0 and piece == 1:
        labels.append([0, 1, 0, 0, 0, 0])
    elif gender == 0 and piece == 2:
        labels.append([0, 0, 1, 0, 0, 0])
    elif gender == 1 and piece == 0:
        labels.append([0, 0, 0, 1, 0, 0])
    elif gender == 1 and piece == 1:
        labels.append([0, 0, 0, 0, 1, 0])
    elif gender == 1 and piece == 2:
        labels.append([0, 0, 0, 0, 0, 1])
    if gender != 2 and piece != 3:                
        labeled_data.append(value)
```

```python
class ClothingDataset(Dataset):
    def __init__(self, data, labels):
        self.labels = labels
        self.data = data
        
    def __getitem__(self, index):
        imgBytes = self.data[index][b'imgs']
        img = np.uint8(np.asarray(Image.open(BytesIO(imgBytes)).convert('RGB').resize((224,224))))
        
        trans = transforms.ToTensor()
        img = trans(img)
        
        label = self.labels[index]
        label = torch.LongTensor(label)
        
        return img, label

    def __len__(self):
        return len(self.labels)
```

```python
training_data = labeled_data[0:114882]
training_labels = labels[0:114882]
validation_data = labeled_data[114882:147706]
validation_labels = labels[114882:147706]
```

```python
'''
MAKE DATA ITERABLE
'''
params = {'batch_size': 64,
          'shuffle': True,
         'num_workers': 6}

training_set = ClothingDataset(training_data, training_labels)
training_generator = DataLoader(training_set, **params)

params['shuffle'] = False
validation_set = ClothingDataset(validation_data, validation_labels)
validation_generator = DataLoader(validation_set, **params)
```

```python
# CUDA for PyTorch
use_cuda = torch.cuda.is_available()
device = torch.device("cuda:0" if use_cuda else "cpu")
```

```python
'''
Model is already defined. Lets instantiate it from torchvision
'''
resnet18 = models.resnet18(num_classes=6).to(device)
```

```python
'''
Instantiate loss class
'''
criterion = nn.CrossEntropyLoss()
```

```python
'''
Instantiate optimizer class
'''
learning_rate = 0.01

optimizer = torch.optim.SGD(resnet18.parameters(), lr=learning_rate)  
```

```python
n_iters = 15387 # for 6 epochs
num_epochs = n_iters / (len(training_set) / 64)
num_epochs = int(num_epochs)
num_epochs
```

```python
'''
Train the model
'''
iter = 0
for epoch in range(num_epochs):
    for images, labels in training_generator:
        # Load images with gradient accumulation capabilities
        #images = images.view(-1, 224*224).requires_grad_()
        images = images.to(device)
        
        labels = labels.to(device)
        
        # Clear gradients w.r.t. parameters
        optimizer.zero_grad()

        # Forward pass to get output/logits
        outputs = resnet18(images)
        
        #print(labels)
        #print(outputs)
        # Calculate Loss: softmax --> cross entropy loss
        labels = torch.max(labels, 1)[1]
        loss = criterion(outputs, labels)

        # Getting gradients w.r.t. parameters
        loss.backward()

        # Updating parameters
        optimizer.step()

        iter += 1

        if iter % 1538 == 0:
            # Calculate Accuracy
            correct = 0
            total = 0
            # Iterate through test dataset
            for images, labels in validation_generator:
                # Load images with gradient accumulation capabilities
                #images = images.view(-1, 224*224).requires_grad_()
                images = images.to(device)
                labels = labels.to(device)
                labels = torch.max(labels, 1)[1]
                # Forward pass only to get logits/output
                outputs = resnet18(images)
                
                # Get predictions from the maximum value
                _, preds = torch.max(outputs, 1)

                # Total number of labels
                total += labels.size(0)

                # Total correct predictions
                correct += (preds == labels).sum()

            accuracy = 100 * correct / total

            # Print Loss
            print('Iteration: {}. Loss: {}. Accuracy: {}'.format(iter, loss.item(), accuracy))
```
